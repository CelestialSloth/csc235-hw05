# https://plotly.com/python/violin/

```python
import plotly.graph_objects as go
import pandas as pd
```

```python
def extract_hour(timestamp):
    return timestamp[5:13]
```

```python
damage_report_df = pd.read_csv('hw05-data/Damage Reports/mc1-reports-data.csv')
damage_report_df['hour'] = damage_report_df['time'].apply(extract_hour)
damage_report_df['time'] = pd.to_datetime(damage_report_df['time'])
```

```python
fig = go.Figure()
hours = damage_report_df['hour'].unique()
for hour in hours:
    fig.add_trace(
        go.Box(
            x=damage_report_df['hour'][damage_report_df['hour']==hour],
            y=damage_report_df['shake_intensity'][damage_report_df['hour'] == hour],
            name=hour,
        )
    )

fig.show()
```
