```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
social_media_df = pd.read_csv('hw05-data/YInt Social Media Data/YInt.csv')
social_media_df['time'] = pd.to_datetime(social_media_df['time'])
```

```python
earthquake_words = ['quake', 'shak']
```

```python
earthquake_msg_df = social_media_df[social_media_df['message'].str.contains(r'', na=False)]
earthquake_msg_df = pd.DataFrame({'count' : earthquake_msg_df.groupby([pd.Grouper(key='time', freq='h'), 'location']).size()}).reset_index()
```

```python
fig = px.line(earthquake_msg_df, x='time', y='count', color='location')
fig.show()
```
